```python
import math

class Circle:
    b: int = 3
    def __init__(self, radius):
        self.radius = radius
        print(self.b)
        self.b = 4

    def area(self):
        return math.pi * self.radius ** 2
    
    def is_bigger_than(self, other: 'Circle'):
        return self.area() > other.area()
    
circle = Circle(3)
print(f'''circle.radius = {circle.radius}''')
print(f'''circle.b = {circle.b}''')
print(circle.area())
```

### 函数

- 构造函数 \_\_init\_\_(self, *args, **kwargs)
- 与js，dart 一样，**不支持方法重载**, 原因在于lookup chain
- 成员函数
    - 第一个参数为实例对象的函数，一般命名self
    - 如果在成员函数体中发现没有使用self,那么考虑class method, static method
    - instance.func()调用，或Class.func(instance)
- class methods
    - 第一个参数为Class对象的函数，一般命名cls
    - @classmethod
    - 一般用作为对象提供多个构造函数，和dart中的命名构造函数一样
    - 用作构造函数时，与kotlin相同，调用主构造函数，也就是\_\_init\_\_
- static methods
    - 定义在class内部的普通函数，没有实例对象/Class对象当作第一个参数, 同java的static方法
    - @staticmethod
    - 可以用Class对象或实例对象调用
- [Instance, Class, and Static Methods](https://realpython.com/instance-class-and-static-methods-demystified/)


```python
class ThreePoint:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    
    @classmethod 
    def from_sequence(cls, sequence):
        return cls(sequence[0], sequence[1], sequence[2])

    def distance(self, other: 'ThreePoint'):
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2 + (self.z - other.z) ** 2) ** 0.5

    def __str__(self) -> str:
        return f'({self.x}, {self.y}, {self.z})'
    
    @staticmethod
    def show_intro_message(name):
        print(f"Hey {name}! This is your 3D Point!")

p1 = ThreePoint(1, 2, 3)
p2 = ThreePoint(4, 5, 6)
print(p1.distance(p2))

l = [1, 2, 3]
p3 = ThreePoint.from_sequence(l)
print(p3)

ThreePoint.show_intro_message('John')
p3.show_intro_message('Josh')
```

### dunder methods & protocols

- dunder methods(**d**ouble **under**score)
    - \_\_init\_\_，构造函数
    - \_\_str\_\_, 相当于toString(), 可以用str(), print()中使用
    - \_\_repr\_\_, 一个能够重构这个对象的正式的字符串表示(formal string representation)，就比如java bean的toJsonString()就是一种java bean的repr表示，repr()调用

- protocols

| Protocol   | Provided Feature                                     | Special Methods                                                          |
| ---------- | ---------------------------------------------------- | ------------------------------------------------------------------------ |
| Iterator   | Allows you to create iterator objects                | .\_\_iter\_\_() and .\_\_next\_\_()                                              |
| Iterable   | Makes your objects iterable                          | .\_\_iter\_\_()                                                              |
| Descriptor | Lets you write managed attributes                    | .\_\_get\_\_() and optionally .\_\_set\_\_(), .\_\_delete\_\_(), and .\_\_set_name\_\_() |
| Context    | manager	Enables an object to work on with statements | .\_\_enter\_\_() and .\_\_exit\_\_()   

```python
class ThreePoint:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __iter__(self):
        yield from (self.x, self.y, self.z)
    
    def __str__(self) -> str:
        return f"({self.x}, {self.y}, {self.z})"
    
    def __repr__(self) -> str:
        return f"ThreePoint({self.x}, {self.y}, {self.z})"
    
p = ThreePoint(1, 2, 3)
print(p)
print(str(p))
print(repr(p))
l = list(p)
print(l)
```

### 可见性

- public: 普通的命名方法: radius, area()
- private: 命名前加下划线: _radius, _area()
- _radius, _area()**约定为private, 但外部调用没有限制，不要使用**
- 双下划线，方法，变量完全被隐藏，__radius, __area()，外部无法调用，内部实际上是做了一个命名替换，_Circle__radius, _Circle__area()

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self._height = height
    
    def get_width(self):
        return self.width

    def _get_height(self):
        return self._height
    
    def __get_area(self):
        return self.width * self._height
    
box = Rectangle(10, 20)
print(box.get_width())
print(box._get_height())
print(box._Rectangle__get_area())
# print(box.__get_area())
```

### setter & getter

- 对于成员变量，python一般不会像java中一样，将变量设为私有，对外暴露set, get方法，而是直接暴露成员变量
- 如果对于成员变量有改动，可以将成员变量转化为property and descriptor-based属性，代码使用保持不变
    - get @property
    - set @attr_name.setter
    - 可以只定义get, set中的一个, 只定义get相当于readOnly
- [Python Descriptors: An Introduction](https://realpython.com/python-descriptors/)
    - 给成员属性增加一些函数样式的功能
    - \_\_get\_\_
    - \_\_set\_\_
    - @LazyProperty

```python
import math

# property 属性
class Circle:
    def __init__(self, radius):
        self.radius = radius

    @property
    def radius(self):
        return self._radius
    
    @radius.setter
    def radius(self, value):
        if (not isinstance(value, int | float) or value < 0):
            raise ValueError("Radius must be a positive real number")
        self._radius = value

    def area(self):
        return math.pi * self.radius ** 2

c = Circle(10)
print(c.radius)
print(c._radius)
print(c.area())

c2 = Circle(-10)
```

```python
import math

class PositiveNumber:
    def __set_name__(self, owner, name):
        self.name = name
    
    def __get__(self, instance, owner):
        return instance.__dict__[self.name]
    
    def __set__(self, instance, value):
        if value < 0:
            raise ValueError(f"{self.name} must be positive")
        instance.__dict__[self.name] = value


class Circle:
    radius = PositiveNumber()
    
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2
    

class Square:
    side = PositiveNumber()
    
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side ** 2

c1 = Circle(10)
print(c1.radius)
print(c1.area())

s1 = Square(10)
print(s1.side)
print(s1.area())
```

### 继承

- 支持多重继承
- super()
- 多重继承查找顺序
    - 当前类
    - 最左边的父类
    - 多重继承的话，从左向右父类
    - 父类的父类
- mixin class
    - mixin class不是种类型，它也不是用来继承的，表示提供一种功能
    - 因为多重继承的顺序，一般mixin class放在最前面
- 常见的类关系
    - 继承
    - 组合
    - 代理
    - 依赖注入

```python
class Shape:
    def __init__(self, name):
        self.name = name

    def extended_info(self):
        print("This is a shape")

    def override_info(self):
        print("This is a shape")
    
class Circle(Shape):
    def __init__(self, name, radius):
        super().__init__(name)
        self.radius = radius
    
    def extended_info(self):
        super().extended_info()
        print("This is a circle")
    
    def override_info(self):
        print("This is a circle")

circle = Circle("Circle", 5)
print(circle.name)
print(circle.radius)
print()
circle.extended_info()
print()
circle.override_info()
```

```python
import json

class Shape:
    def __init__(self, name):
        self.name = name

# mixin class
class SerializerMixin:
    def to_json(self):
        return json.dumps(self.__dict__)
    
# mixin class 放在最左边
class Circle(SerializerMixin,  Shape):
    def __init__(self, name, radius):
        super().__init__(name)  
        self.radius = radius


c = Circle("Circle", 10)
print(c.to_json())
```

### ABC(Abstract Base classes)

- 与java中的接口一样，子类必须实现所有方法

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Square(Shape):
    def __init__(self, side):
        self.__side = side

    def area(self):
        return self.__side**2

    def perimeter(self):
        return self.__side*4
    
class Circle(Shape):
    def __init__(self, radius):
        self.__radius = radius

    def area(self):
        return math.pi*self.__radius**2

s = Square(5)
print(s.area())
print(s.perimeter())

c = Circle(5)
print(c.area())
print(c.perimeter())
```
